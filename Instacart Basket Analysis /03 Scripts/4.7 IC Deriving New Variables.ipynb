## Using If-Statements

```python
# Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

## If-Statements with the loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### STEP 4: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Checking frequency of daily orders in 'order_hour_of_day' column

df_ords_prods_merged['order_hour_of_day'].value_counts()
```

#### As the client wants three time periods, we can divide the above in three equal sections of eight hours each. 'Most orders' will consist of hours 10, 11, 14, 15, 13, 12, 16 and 9. 'Avarage orders', will consist of hours 17, 8, 18, 19, 20, 7, 21 and 22. 'Fewest orders' will consist of the remaining 8 hours, 23, 6, 0, 1, 5, 2, 4 and 3.

```python
# Creating new if statement

period_of_time = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    period_of_time.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    period_of_time.append("Fewest orders")
  else:
    period_of_time.append("Average orders")
```

```python
# Add result to dataframe as "busiest_period_of_day" column

df_ords_prods_merged['busiest_period_of_day'] = period_of_time
```
