### STEP 3: You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the "order_hour_of_day" column. Describe what the histogram demonstrates.

```python
hist4 = instacart_sub['order_hour_of_day'].plot.hist(bins = 75)
plt.title('Distribution of Orders by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Count of Orders')
```

### STEP 7: Explore whether there's a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function. Describe what the scatterplot shows.
